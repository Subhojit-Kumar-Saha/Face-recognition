# Face Data Storing


```python
import numpy as np
import pandas as pd
import cv2
```

```python
d = pd.DataFrame()
#df = pd.read_csv("FaceData.csv")
```

```python
f=[]
skip=0
name = np.str(input("Enter your name"))
webcam = cv2.VideoCapture(0)
while True:
    ret,frame = webcam.read()
        
    facecascade = cv2.CascadeClassifier("face.xml")
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces = facecascade.detectMultiScale(gray,1.5,5)
    if(len(faces)==0):
        continue
    faces = sorted(faces,key = lambda x:x[2]*x[3])
    
    for (x,y,w,h) in faces[-1:]:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
        face = frame[y:y+h,x:x+w]
        skip+=1
        if skip%10==0 and face.reshape((1,-1)).shape[1]<155000:
            f.append(face.reshape((1,-1)))
            
        
    
    cv2.imshow("DATA SAVING",frame)
    
    if(cv2.waitKey(1) & 0XFF == ord('q')):
        break

webcam.release()
cv2.destroyAllWindows()
f = np.asarray(f)
f = f.reshape((-1,f.shape[2]))
dx = pd.DataFrame(f)
l = [name for x in range(dx.shape[0])]
l = np.asarray(l)
l = l.reshape((-1,1))
z = np.hstack((l,dx))
z= pd.DataFrame(z)
d=d.append(z)
i = pd.Index(range(d.shape[0]))
d.index = range(d.shape[0])
```

# Face Recognition


```python
df = pd.read_csv("FaceData.csv")
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
clf = KNeighborsClassifier()
```

```python
X = df.iloc[:,1:]
Y = df.iloc[:,0]
```

```python
clf.fit(X,Y)
```

```python
import cv2
```

```python
webcam = cv2.VideoCapture(0)

while True:
    ret,frame = webcam.read()    
    facecascade = cv2.CascadeClassifier("face.xml")
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces = facecascade.detectMultiScale(gray,1.5,5)
    if(len(faces)==0):
        continue
    faces = sorted(faces,key = lambda x:x[2]*x[3])
    
    for (x,y,w,h) in faces[-1:]:
        
        face = frame[y:y+h,x:x+w]
        face = face.reshape((1,-1))
        if face.reshape((1,-1)).shape[1]==69312:
            label = clf.predict(face)
            cv2.putText(frame,label[0],(x,y-10),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,255),2,cv2.LINE_AA)
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
            
        
    
    cv2.imshow("DETECT",frame)
    
    if(cv2.waitKey(1) & 0XFF == ord('q')):
        break

webcam.release()
cv2.destroyAllWindows()
```
